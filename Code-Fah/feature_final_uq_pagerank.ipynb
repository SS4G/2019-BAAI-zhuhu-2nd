```python
import pandas as pd
```

```python
train=pd.read_csv('invite_info.txt',sep='\s+',names=['qid','uid','time','target'])
test=pd.read_csv('invite_info_evaluate_1.txt',sep='\s+',names=['qid','uid','time'])
```

```python
trn_uid=train['uid'].tolist()
trn_qid=train['qid'].tolist()
te_uid=test['uid'].tolist()
te_qid=test['qid'].tolist()
```

```python
from tqdm import tqdm
tqdm.pandas()
```

```python
def generate_qid_graph_table(uid,qid):
    hash_key1 =uid
    hash_key2 =qid

    qid_graph.setdefault(hash_key1, []).append(hash_key2)
    qid_graph.setdefault(hash_key2, []).append(hash_key1)
```

```python
qid_graph = {}
for i in tqdm(range(len(trn_uid))):
    generate_qid_graph_table(trn_uid[i],trn_qid[i])
for i in tqdm(range(len(te_uid))):
    generate_qid_graph_table(te_uid[i],te_qid[i])
```

```python
def pagerank():
    MAX_ITER = 40
    d = 0.85

    pagerank_dict = {i: 1 / len(qid_graph) for i in qid_graph}
    num_nodes = len(pagerank_dict)

    for iter in tqdm(range(0, MAX_ITER)):
        print(iter)
        for node in qid_graph:
            local_pr = 0

            for neighbor in qid_graph[node]:
                local_pr += pagerank_dict[neighbor] / len(qid_graph[neighbor])

            pagerank_dict[node] = (1 - d) / num_nodes + d * local_pr

    return pagerank_dict
```

```python
i=0
pagerank_dict = pagerank()
```

```python
def get_pagerank_fun(df):
    result=[]
    def get_pagerank_value(row):
        result.append([pagerank_dict[row["qid"]],pagerank_dict[row["uid"]]])
    pr_feats=df.progress_apply(get_pagerank_value, axis=1)
    return pd.DataFrame(result)
```

```python
feature_trn=get_pagerank_fun(train)
```

```python
feature_te=get_pagerank_fun(test)
```

```python
feature=pd.concat([feature_trn,feature_te])
```

```python
feature.columns=['pr_q','pr_a']
```

```python
feature=feature*1000000
```
