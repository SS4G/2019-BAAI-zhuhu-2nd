```python
import pandas as pd
import time
import numpy as np
```

```python
train=pd.read_csv('invite_info.txt',sep='\s+',names=['qid','uid','time','target'])
test=pd.read_csv('invite_info_evaluate_1.txt',sep='\s+',names=['qid','uid','time'])
test2=pd.read_csv('invite_info_evaluate_2_0926.txt',sep='\s+',names=['qid','uid','time'])
all_data=pd.concat([train[['qid','uid','time']],test2,test])
```

```python
from tqdm import tqdm
tqdm.pandas()
```

```python
all_data['day']=all_data['time'].progress_apply(lambda x:x[1:].split('-H')[0])
```

```python
def caul_time(x):
    time=x[1:].split('-H')
    return int(time[0])*24+int(time[1])
```

```python
all_data['time']=all_data['time'].progress_apply(lambda x:caul_time(x))
```

```python
def time_diff_stat(ser):  #code from 陈雄君 陈垂泽 雅典娜杯
    ts = ser.values
    ts_diff = np.diff(ts)
    if ts_diff.shape[0] < 2:
        return 0, 0, 0, 0
    else:
        return ts_diff.min(), ts_diff.max(), ts_diff.mean(), ts_diff.std()
```

```python
all_data.sort_values(by=['time'], inplace=True)
```

```python
feature = all_data.groupby(['uid'])['time'].progress_apply(time_diff_stat).reset_index()
```

```python
feature=pd.concat([feature,pd.DataFrame(feature['time'].tolist())],axis=1)
```

```python
feature.drop(['time'],axis=1,inplace=True)
feature.columns=['uid','intime_gap_min','intime_gap_max','intime_gap_mean','intime_gap_std']
```

```python
feature=feature.reset_index(drop=True)
```

```python
feature.to_hdf('feature_intime_gap2.h5', key='data')
```

```python
featureq = all_data.groupby(['qid'])['time'].progress_apply(time_diff_stat).reset_index()
```

```python
featureq=pd.concat([featureq,pd.DataFrame(featureq['time'].tolist())],axis=1)
```

```python
featureq.drop(['time'],axis=1,inplace=True)
featureq.columns=['qid','qintime_gap_min','qintime_gap_max','qintime_gap_mean','qintime_gap_std']
```

```python
featureq=featureq.reset_index(drop=True)
```

```python
featureq.to_hdf('feature_qintime_gap2.h5', key='data')
```
