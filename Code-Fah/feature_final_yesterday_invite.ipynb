```python
import pandas as pd
import time
from tqdm import tqdm
tqdm.pandas()
time_start=time.time()
```

```python
train=pd.read_csv('invite_info.txt',sep='\s+',names=['qid','uid','time','target'])
test2=pd.read_csv('invite_info_evaluate_2_0926.txt',sep='\s+',names=['qid','uid','time'])
test=pd.read_csv('invite_info_evaluate_1.txt',sep='\s+',names=['qid','uid','time'])
```

```python
#计算问题时间 和问题时间减去邀请时间
```

```python
train['invite_day']=train['time'].progress_apply(lambda x:x[1:].split('-H')[0])
test['invite_day']=test['time'].progress_apply(lambda x:x[1:].split('-H')[0])
```

```python
test2['invite_day']=test2['time'].progress_apply(lambda x:x[1:].split('-H')[0])
```

```python
all_data=pd.concat([train[['qid','uid','time','invite_day']],test2,test])
```

```python
feature_q=all_data.groupby(['qid','invite_day']).agg({'uid':'count'})
```

```python
feature_q=feature_q.reset_index()
feature_q.columns=['qid','invite_day','q_yesterday_invite']
```

```python
feature_u=all_data.groupby(['uid','invite_day']).agg({'qid':'count'}).reset_index()
feature_u.columns=['uid','invite_day','u_yesterday_invite']
```

```python
feature_u.head()
```

```python
feature_u['invite_day']=feature_u['invite_day'].astype(int)+1
feature_q['invite_day']=feature_q['invite_day'].astype(int)+1
```

```python
feature_u['invite_day']=feature_u['invite_day'].astype(str)
feature_q['invite_day']=feature_q['invite_day'].astype(str)
all_data['invite_day']=all_data['invite_day'].astype(str)
```

```python
all_data=pd.merge(all_data,feature_u,how='left',on=['uid','invite_day'])
all_data=pd.merge(all_data,feature_q,how='left',on=['qid','invite_day'])
```

```python
all_data=all_data.drop_duplicates(['qid','uid'])
```

```python
all_data=all_data.reset_index(drop=True)
```
